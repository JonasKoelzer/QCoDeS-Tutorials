# 2D Loop test

In this tutorial you will learn how to set up and live plot a 2D Loop experiment. This is in principle not so much different from a 1D loop, but still it is nice to learn how it is done. Usefull applications could for example be a transport measurement, where you want to sweep two different gates (like a top and a backgate) or the bias and the gate. We start by importing our usual packages:

```python
import qcodes as qc

from qcodes.tests.instrument_mocks import DummyInstrument
from qcodes.instrument.parameter import Parameter
from qcodes.instrument.specialized_parameters import ElapsedTimeParameter
from qcodes.loops import Loop
```

In the next step we define the instruments such as the parameters. The only difference to what we did before is that we now define an additional instrument, since we need two parameters to sweep for a 2d loop:

```python
src1 = DummyInstrument(name='src1', gates=['v1', 'v2'])
src2 = DummyInstrument(name='src2', gates=['v3', 'v4'])

t = ElapsedTimeParameter('t')
V_appl_1 = Parameter('V', label='V_appl_1', unit='V', get_cmd=src1.v1.get, set_cmd=src1.v1.set)
V_appl_2 = Parameter('V', label='V_appl_2', unit='V', get_cmd=src2.v3.get, set_cmd=src2.v3.set)
I_meas = Parameter('I', label='I_meas', unit='A', get_cmd=lambda: float((src1.v1()+src2.v3())/1000))
```

Now we have to set up our loop. This is basically done by adding a second loop step to the first one.
This kind of a loop will later result in a 2D scheme, where V_appl_2 is swept in the inner loop and V_appl_1 is swept in the outer loop. 

```python
t.reset_clock()
loop = Loop(V_appl_1.sweep(-1,1,0.01), delay = 0.001).loop(V_appl_2.sweep(-1,1,0.01), delay = 0.001).each(V_appl_1,V_appl_2,I_meas,t)
```

As a last step we only need to start the loop (hte same way we would also do this in the 1D scenario) and qcodes automatically manages the sweep in the two loops such as the plot generartion. Here is an exmplae how a live plot would be generated:

```python
%matplotlib inline
from qcodes.plots.pyqtgraph import QtPlot

data = loop.get_data_set(name='sweep')
plot = QtPlot()
plot.add(data.I)
_ = loop.with_bg_task(plot.update).run()
```

An this is what the final plot looks like. In case of a live plot you also have the possibility to dynamically adjust the stepping of the color bar during the measurement so that you can highlight steps or features in your dataset by setting the triangle on the right side of the color bar.

```python
plot
```

In order to save this plot we just have to do this:

```python
plot.save('IV_2D_sweep_plot.png')
```
