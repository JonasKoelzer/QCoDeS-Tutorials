# Run a 1D Loop

The first thing that we want to do is to perform a simplet sweep experiment, meaning we want to change one parameter and want to measure at least one parameter. 
In order to do so we first have to import the required packages:

```python
import qcodes as qc

from qcodes.tests.instrument_mocks import DummyInstrument
from qcodes.instrument.parameter import Parameter
from qcodes.instrument.specialized_parameters import ElapsedTimeParameter
from qcodes.loops import Loop
```

Next we have to define our instruments and parameters:

```python
src1 = DummyInstrument(name='src1', gates=['v1', 'v2'])
```

```python
t = ElapsedTimeParameter('t')
V_appl = Parameter('V', label='V_appl', unit='V', get_cmd=src1.v1.get, set_cmd=src1.v1.set)
I_meas = Parameter('I', label='I_meas', unit='A', get_cmd=lambda: float(src1.v1()/1000))
```

Now we can define the experiment. This is done by using the loop function in QCoDeS. In this case we want to sweep the voltage applied and measure the current and the time:

```python
t.reset_clock()
loop = Loop(V_appl.sweep(-1,1,0.01), delay = 0.01).each(V_appl,I_meas,t)
```

Finally we run the measurement by simply executing the following command. We receive a set of information about start/end date of the experiment such as the shape of the data that has been aquired:

```python
loop.run()
```

# Plotting Loop Data

It is always helpful to also plot the data. This is done by using the plot function that comes with the loop experiment:

```python
%matplotlib inline
from qcodes.plots.pyqtgraph import QtPlot

loop = Loop(V_appl.sweep(-1,1,0.01), delay = 0.01).each(V_appl,I_meas,t)
data = loop.get_data_set(name='sweep')
plot = QtPlot()
plot.add(data.I)
loop.run()
plot.update()
```

It is possible to plot the data inside the jupyter notebook:

```python
plot
```

### Additional note
Because there have been some questions about this: It is very useful to store the plot that we just created for quick presentation of the raw data. This is done by as follows:

```python
plot.save('IV_sweep_plot.png')
```

Plots give us a quick feedback about the measurement that we are running. The fastest way of getting this feedback is by live plotting the data aquired. The loop functions comes with a built-in live plotting function which is implemented as follows:

```python
loop = Loop(V_appl.sweep(-1,1,0.01), delay = 0.1).each(V_appl,I_meas,t)
data = loop.get_data_set(name='sweep')
plot = QtPlot()
plot.add(data.I)
_ = loop.with_bg_task(plot.update).run()
```

# Storing Data 

We all know the problem: Some time after the measurement you are searching for a particular file of data that you measured. You remeber exactly the setup you used and you also remember the time at which you did it, but where did the file go?
In order to aviod this file-mess QCoDeS comes with a native database integration, which is implemented into a loop in just a few steps.
It automatically formats/stores and manages the data for you in an sqlite database. You just have to tell QCoDeS the database location and everything else is taken care of. This means: You can not accidentally overwrite an experiment (every experiment has a uniquie ID). Some quick tips: Use a storage drive that is protected by some kind of backup system (like raid), use one db for each setup, since only one script can write into the db at a time. Storing the data per setup helps you on a long term to collect a lot of data in one place.
You start by defining the database location such as:

```python
from qcodes import initialise_database, load_or_create_experiment
from qcodes.data.location import FormatLocation
from qcodes.dataset.legacy_import import import_dat_file
from qcodes.data.io import DiskIO

exp_name = 'DC-IV-Sweep'
sample_name = 'Test-Resistor'

qc.config["core"]["db_location"]='./experiments.db'
initialise_database()
exp=load_or_create_experiment(exp_name,sample_name)
```

```python
loop = Loop(V_appl.sweep(-1,1,0.01), delay = 0.001).each(V_appl,I_meas,t)
```

This is how you store the data aquired by the loop in a .dat file 

```python
loc_fmt='./data/{date}/#{counter}_{name}_{sample}'
rcd={'name':exp_name, 'sample': sample_name}
loc_provider = FormatLocation(fmt=loc_fmt)
loc = loc_provider(DiskIO("."),record=rcd)
data = loop.get_data_set(location=loc)
```

```python
loop.run()
```

And here is how you import that .dat file's data into the database:

```python
import_dat_file(loc,exp)
```
